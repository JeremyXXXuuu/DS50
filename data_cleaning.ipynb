```python
def desc_getter(stringg):
    game_type=""
    row=stringg
    row=re.sub(r'[^\x00-\x7f]',r'', row)
    if(len(row)>10):
        if(row[11]=='G'):
            game_type="game"
        elif(row[11]=='C'):
            game_type="content"
        elif(row[11]=='b'):
            game_type="bundle"
    return game_type
```

```python
tags=[0]*len(steam)
```

```python
for indexs in steam.index:
    tags[indexs] = steam["popu_tags"][indexs]
```

```python
def idx2appid(idx):
    appid = int(steam.loc[idx,"appid"])
    return appid
```

```python
def get_score(game_index):
    score = [0]*length
    for i in range(len(tags[game_index])):
        for indexs in steam.index:
            if(indexs != game_index):
                for j in range(len(tags[indexs])):
                    if tags[game_index][i] == tags[indexs][j]:
                        score[indexs] = score[indexs]+1
    return score
```

```python
def get_index2(game_index):
    game_tab=[0]*10
    score=get_score(game_index)
    re1 = heapq.nlargest(10, score)
    for i in range(len(re1)):
        re2=[0]*2
        re2[1] = re1[i]
        idx = score.index(re1[i])
        re2[0] = idx2appid(idx)
        if i>0 and re2[0] == game_tab[i-1][0]:
            score[idx]=0
            idx = score.index(re1[i])
            re2[0] = idx2appid(idx)
        game_tab[i]=re2
    return game_tab
```
